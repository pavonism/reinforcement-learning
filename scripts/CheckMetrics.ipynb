```python
import pandas as pd
import matplotlib.pyplot as plt

CHECKPOINT_NAME = r'ppo/5mln-larger-smaller-entropy/training_log_2024-12-20_11-16-20.txt'
df = pd.read_csv(f'../checkpoints/{CHECKPOINT_NAME}', index_col=False)

df.groupby('episode').sum()['reward'].rolling(window=200, center=True).mean().plot(y='reward')
df.groupby('episode').sum()['reward'].expanding().mean().plot(y='reward')
plt.legend(['Rolling mean', 'Expanding mean'])

df.groupby('episode').sum().plot(y='reward')
plt.legend(['Reward'])
# df.groupby('episode')[['episode', 'loss']].mean().plot(y='loss', logy=True)
episodes_in_metrics = df.groupby("episode").count()
```
